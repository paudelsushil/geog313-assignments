```python
 # Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating the array x
x = np.arange(-100, 101, 0.5)
```

```python
# Generating the arrays y and z
y = np.cos(x)
z = np.sin(x)
```

```python
# Generating the arrays y and z
y = np.cos(x)
z = np.sin(x)
```

```python
# Plot 
plt.plot(x, y, label='cos(x)', color='orange')
plt.plot(x, z, label='sin(x)', color='blue')

# Add a legend
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))

# Add x and y axis labels
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.grid(True)
plt.title("Assignment 2")
plt.show()
```
